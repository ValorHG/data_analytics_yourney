```python
import pandas as pd
import numpy as np
from sklearn.cluster import AffinityPropagation
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Assuming osiPCA is already created using PCA with 5 components:
# (Replace 'osiPCA' with your actual PCA-transformed data if needed)

# Step 1: Select only the first 1500 items from the PCA-transformed data
AffPropClus = osiPCA[:3000, :2]  # Use the first 1500 items and the first 2 PCA components for clustering and visualization

# Step 2: Fit Affinity Propagation clustering with updated parameters
clustering = AffinityPropagation(
    random_state=5, 
    damping=0.75,  # Increased damping for stability
    max_iter=10000,  # Allow more iterations for convergence
    preference=-500  # Adjust preference to control clusters
).fit(AffPropClus)

# Step 3: Get the cluster labels and cluster centers
affinity_labels = clustering.labels_
cluster_centers = clustering.cluster_centers_

# Step 4: Set up the color map
unique_labels = np.unique(affinity_labels)
colors = cm.get_cmap('hsv', len(unique_labels))

# Step 5: Plot the clustering result for the first 1500 items
plt.figure(figsize=(10, 8))
for i, label in enumerate(unique_labels):
    cluster_points = AffPropClus[affinity_labels == label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                color=colors(i), label=f'Cluster {label}')

# Step 6: Plot cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
            color='black', marker='X', s=200, label='Cluster Centers')

# Step 7: Add labels and title
plt.title('Affinity Propagation Clustering (First 1500 Items of PCA Transformed Data)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Step 1: Select 5000 items from the transformed data
DbsClusData = osiPCA[:5000, :2]  # Use the 5000 items and the 2 PCA components for clustering

# Step 2: Set DBSCAN parameters (You can edit these variables later)
eps = 0.7  # Maximum distance between two samples to be considered in the same neighborhood
min_samples = 50  # Minimum number of samples required to form a core point

# Step 3: Fit DBSCAN clustering
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(DbsClusData)

# Step 4: Set up the color map
unique_labels = np.unique(dbscan_labels)
# Updated way to get colormap
colors = plt.colormaps['hsv'](np.linspace(0, 1, len(unique_labels)))

# Step 5: Plot the clustering result
plt.figure(figsize=(10, 8))
for i, label in enumerate(unique_labels):
    cluster_points = DbsClusData[dbscan_labels == label]
    if label == -1:
        # Black color for noise points (label = -1 in DBSCAN)
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color='black', label='Noise')
    else:
        # Different color for each cluster
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[i], label=f'Cluster {label}')

# Step 6: Add labels and title
plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import Birch
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming osiPCA is already created using PCA with 2 components:
# (Replace 'osiPCA' with your actual PCA-transformed data if needed)

# Step 1: Select only the first 5000 items from the PCA-transformed data
BirchClusData = osiPCA[:5000, :2]  # Rename X to birch_data for clarity

# Step 2: Set Birch parameters (You can edit these variables later)
threshold = 0.5  # The radius of the subclusters obtained by merging
branching_factor = 50  # Maximum number of subclusters in each node of the tree

# Step 3: Fit Birch clustering
birch = Birch(threshold=threshold, branching_factor=branching_factor)
birch_labels = birch.fit_predict(BirchClusData)  # Rename labels to birch_labels

# Step 5: Generate bright colors for each cluster using Seaborn's bright palette
num_clusters = len(np.unique(birch_labels))  # Number of unique clusters
palette = sns.color_palette("bright", num_clusters)  # Use Seaborn bright palette for distinct colors

# Step 6: Plot the clustering result with bright colors
plt.figure(figsize=(10, 8))
for i, label in enumerate(np.unique(birch_labels)):
    cluster_points = BirchClusData[birch_labels == label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                color=palette[i], label=f'Cluster {label}')

# Step 7: Add labels and title
plt.title(f'Birch Clustering (threshold={threshold}, branching_factor={branching_factor})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
# Function to calculate the Euclidean distance between two points
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))

# Function to calculate the silhouette score for a single data point
def silhouette_score_point(AffDataClus, labels, i):
    # Get the label of the point and the cluster it belongs to
    point = AffDataClus[i]
    label = labels[i]

    # Step 1: Calculate a(i) - Intra-cluster distance
    same_cluster = AffDataClus[labels == label]
    if len(same_cluster) == 1:
        # Only one point in this cluster, silhouette score is 0
        return 0
    
    a_i = np.mean([euclidean_distance(point, other) for other in same_cluster if not np.array_equal(point, other)])

    # Step 2: Calculate b(i) - Nearest cluster distance
    unique_labels = np.unique(labels)
    b_i = float('inf')  # Set to a large number initially

    for other_label in unique_labels:
        if other_label == label:
            continue  # Skip the current cluster
        
        other_cluster = AffDataClus[labels == other_label]
        avg_distance = np.mean([euclidean_distance(point, other) for other in other_cluster])
        b_i = min(b_i, avg_distance)  # Find the minimum average distance to the nearest cluster

    # Step 3: Calculate the silhouette score for this point
    s_i = (b_i - a_i) / max(a_i, b_i)
    return s_i

# Function to calculate the overall silhouette score for a clustering result
def silhouette_score(AffDataClus, labels):
    scores = []
    for i in range(len(AffDataClus)):
        score = silhouette_score_point(AffDataClus, labels, i)
        scores.append(score)
    
    return np.mean(scores)

# Step 9: Calculate and print the silhouette score for Affinity Propagation clustering
silhouette = silhouette_score(AffDataClus, labels)
print(f'Affinity Propagation Silhouette Score: {silhouette}')
```

```python
import math

# Function to manually calculate the Euclidean distance between two points
def euclidean_distance(p1, p2):
    return math.sqrt(sum((x1 - x2) ** 2 for x1, x2 in zip(p1, p2)))

# Function to manually calculate the mean
def manual_mean(values):
    return sum(values) / len(values) if values else 0

# Function to calculate the silhouette score for a single data point
def silhouette_score_point(AffDataClus, labels, i):
    # Get the label of the point and the cluster it belongs to
    point = AffDataClus[i]
    label = labels[i]

    # Step 1: Calculate a(i) - Intra-cluster distance
    same_cluster = [AffDataClus[idx] for idx in range(len(AffDataClus)) if labels[idx] == label]
    if len(same_cluster) == 1:
        # Only one point in this cluster, silhouette score is 0
        return 0

    a_i = manual_mean([euclidean_distance(point, other) for other in same_cluster if other is not point])

    # Step 2: Calculate b(i) - Nearest cluster distance
    unique_labels = set(labels)  # Use set to avoid duplicate labels
    b_i = float('inf')  # Set to a large number initially

    for other_label in unique_labels:
        if other_label != label:
            other_cluster = [AffDataClus[idx] for idx in range(len(AffDataClus)) if labels[idx] == other_label]
            avg_distance = manual_mean([euclidean_distance(point, other) for other in other_cluster])
            b_i = min(b_i, avg_distance)  # Find the minimum average distance to the nearest cluster

    # Step 3: Calculate the silhouette score for this point
    s_i = (b_i - a_i) / max(a_i, b_i) if max(a_i, b_i) != 0 else 0
    return s_i

# Function to calculate the overall silhouette score for a clustering result
def silhouette_score(AffDataClus, labels):
    scores = []
    for i in range(len(AffDataClus)):
        score = silhouette_score_point(AffDataClus, labels, i)
        scores.append(score)
    
    return manual_mean(scores)

# Example usage (Assuming AffDataClus and labels are available)
silhouette = silhouette_score(AffDataClus, labels)
print(f'Affinity Propagation Silhouette Score: {silhouette}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import calinski_harabasz_score

# Step 1: Manual Euclidean distance function (without using predefined functions)
def manual_euclidean_distance(p1, p2):
    return sum((x1 - x2) ** 2 for x1, x2 in zip(p1, p2)) ** 0.5

# Step 2: Select 5000 items from the transformed data (PCA-reduced data)
DbsClusData = osiPCA[:5000, :2]  # Use the 5000 items and the first 2 PCA components for clustering

# Step 3: Set DBSCAN parameters (You can adjust these values if needed)
eps = 0.7  # Maximum distance between two samples to be considered in the same neighborhood
min_samples = 50  # Minimum number of samples required to form a core point

# Step 4: Fit DBSCAN clustering with custom Euclidean distance
dbscan = DBSCAN(
    eps=eps,
    min_samples=min_samples,
    metric=manual_euclidean_distance  # Use the custom distance function
)
labels_euclidean = dbscan.fit_predict(DbsClusData)  # Changed 'dbscan_labels' to 'labels_euclidean'

# Step 5: Set up the color map for the clusters
unique_labels = np.unique(labels_euclidean)
colors = plt.colormaps['hsv'](np.linspace(0, 1, len(unique_labels)))  # Get distinct colors

# Step 6: Plot the clustering result
plt.figure(figsize=(10, 8))
for i, label in enumerate(unique_labels):
    cluster_points = DbsClusData[labels_euclidean == label]
    if label == -1:
        # Black color for noise points (label = -1 in DBSCAN)
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color='black', label='Noise')
    else:
        # Different color for each cluster
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[i], label=f'Cluster {label}')

# Step 7: Add labels and title
plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples}) with Custom Euclidean Distance')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import calinski_harabasz_score

# Step 1: Manual Manhattan distance function (without using predefined functions)
def manual_manhattan_distance(p1, p2):
    return sum(abs(x1 - x2) for x1, x2 in zip(p1, p2))

# Step 2: Select 5000 items from the transformed data (PCA-reduced data)
DbsClusData = osiPCA[:5000, :2]  # Use the 5000 items and the first 2 PCA components for clustering

# Step 3: Set DBSCAN parameters (You can adjust these values if needed)
eps = 0.7  # Maximum distance between two samples to be considered in the same neighborhood
min_samples = 50  # Minimum number of samples required to form a core point

# Step 4: Fit DBSCAN clustering with custom Manhattan distance
dbscan = DBSCAN(
    eps=eps,
    min_samples=min_samples,
    metric=manual_manhattan_distance  # Use the custom Manhattan distance function
)
labels_manhattan = dbscan.fit_predict(DbsClusData)  # Changed 'dbscan_labels' to 'labels_manhattan'

# Step 5: Set up the color map for the clusters
unique_labels = np.unique(labels_manhattan)
colors = plt.colormaps['hsv'](np.linspace(0, 1, len(unique_labels)))  # Get distinct colors

# Step 6: Plot the clustering result
plt.figure(figsize=(10, 8))
for i, label in enumerate(unique_labels):
    cluster_points = DbsClusData[labels_manhattan == label]
    if label == -1:
        # Black color for noise points (label = -1 in DBSCAN)
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color='black', label='Noise')
    else:
        # Different color for each cluster
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[i], label=f'Cluster {label}')

# Step 7: Add labels and title
plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples}) with Custom Manhattan Distance')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import calinski_harabasz_score

# Step 1: Manual Cosine distance function (without using predefined functions)
def manual_cosine_distance(p1, p2):
    dot_product = sum(x1 * x2 for x1, x2 in zip(p1, p2))
    norm_p1 = sum(x1 ** 2 for x1 in p1) ** 0.5
    norm_p2 = sum(x2 ** 2 for x2 in p2) ** 0.5
    cosine_similarity = dot_product / (norm_p1 * norm_p2) if norm_p1 != 0 and norm_p2 != 0 else 0
    return 1 - cosine_similarity

# Step 2: Select 5000 items from the transformed data (PCA-reduced data)
DbsClusData = osiPCA[:5000, :2]  # Use the 5000 items and the first 2 PCA components for clustering

# Step 3: Set DBSCAN parameters (You can adjust these values if needed)
eps = 0.05  # Maximum distance between two samples to be considered in the same neighborhood
min_samples = 5  # Minimum number of samples required to form a core point

# Step 4: Fit DBSCAN clustering with custom Cosine distance
dbscan_cosine = DBSCAN(
    eps=eps,
    min_samples=min_samples,
    metric=manual_cosine_distance  # Use the custom Cosine distance function
)
labels_cosine = dbscan_cosine.fit_predict(DbsClusData)  # Use labels_cosine

# Step 5: Set up the color map for the clusters
unique_labels = np.unique(labels_cosine)
colors = plt.colormaps['hsv'](np.linspace(0, 1, len(unique_labels)))  # Get distinct colors

# Step 6: Plot the clustering result
plt.figure(figsize=(10, 8))
for i, label in enumerate(unique_labels):
    cluster_points = DbsClusData[labels_cosine == label]
    if label == -1:
        # Black color for noise points (label = -1 in DBSCAN)
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color='black', label='Noise')
    else:
        # Different color for each cluster
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[i], label=f'Cluster {label}')

# Step 7: Add labels and title
plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples}) with Custom Cosine Distance')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 5: Calculate Calinski-Harabasz Score for each clustering result

ch_score_euclidean = calinski_harabasz_score(DbsClusData, labels_euclidean)
ch_score_manhattan = calinski_harabasz_score(DbsClusData, labels_manhattan)
ch_score_cosine = calinski_harabasz_score(DbsClusData, labels_cosine)

# Step 6: Output the Calinski-Harabasz Scores
print(f'Calinski-Harabasz Score for DBSCAN with Euclidean Distance: {ch_score_euclidean}')
print(f'Calinski-Harabasz Score for DBSCAN with Manhattan Distance: {ch_score_manhattan}')
print(f'Calinski-Harabasz Score for DBSCAN with Cosine Distance: {ch_score_cosine}')
```
