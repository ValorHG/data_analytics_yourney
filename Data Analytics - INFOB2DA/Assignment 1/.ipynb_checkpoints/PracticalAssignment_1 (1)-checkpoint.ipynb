```python
# Task 1: Download and import dataset (5 Points) (1.1)

import pandas as pd

# Reading the CSV file
mmgData = pd.read_csv('mammographic_masses_data.csv')

# Displaying the first few rows of the DataFrame
print(mmgData.to_string())
```

```python
# with the shape fucntion we count the amount of rows and colums

mmgData.shape
```

```python
# with the notna function we removed all the rows that had a NaN in it, meaning that data was lost or either not recorded.
# the all(axis) makes sure all colums have no null values anymore

mmgData_notnan = mmgData[mmgData[["Density", "Margin", "Shape", "BA", "Age", "Severity"]].notna().all(axis=1)]

mmgData_notnan = mmgData_notnan[mmgData_notnan['BA'] != 0]

print(mmgData_notnan.to_string())
```

```python
mmgData_notnan.shape
```

```python
# As can be seen here we lost 131 rows of data. But looking at that we have 829 out of 961 remaining that is only a 13.6% loss of data.
```

```python
""" 
So following now we are going to look at the kind of summaries we can make, and try to find conclusions for this data.

Next we made 3 correlations, one is between Age and Severity. To see if
a certain age range has more servere cases compared to other age ranges.

Next we can look if BI-RADS(BA) has any relationship with the Margin. maybe saying a certain mass margin has a higher chance to be malignant.

finally Density and BI-RADS(BA), see if there is a relation there.
"""
```

```python
# here we use panda's cut funtion to define the ranges and 20 to 90 and the taking steps by 10
mmgData_notnan.loc[:, 'Age_Range'] = pd.cut(mmgData_notnan['Age'], bins=range(10, 110, 10), right=False)

# Get the age ranges ina cloumm by group and calculate the average severity for each age range
age_severity = mmgData_notnan.groupby('Age_Range')['Severity'].mean().reset_index()

print(age_severity.to_string())
```

```python
import plotly.express as px

# Makes strings of the interavls of the age range so plotly can read it
age_severity['Age_Range'] = age_severity['Age_Range'].astype(str)

# Creates the bar plot 
mmg_Bar = px.bar(age_severity, x='Age_Range', y='Severity', title='Average Severity by Age Range')

mmg_Bar.show()
```

```python
mmgData_notnan_filtered = mmgData_notnan[(mmgData_notnan['BA'] >= 1) & (mmgData_notnan['BA'] <= 6)]

grouped_data = mmgData_notnan_filtered.groupby(['Density', 'BA']).size().reset_index(name='count')

grouped_data['percentage'] = grouped_data['count'] / grouped_data.groupby('Density')['count'].transform('sum') * 100

fig = px.histogram(grouped_data, x='Density', y='percentage', color='BA', 
                   barmode='group', 
                   title="Percentage Distribution of BI-RADS Scores by Breast Density",
                   labels={"Density": "Breast Density", "BA": "BI-RADS Score", "percentage": "Percentage (%)"},
                   category_orders={"Density": ['1', '2', '3', '4'], "BA": ['1', '2', '3', '4', '5', '6']})

fig.update_layout(
    xaxis_title="Breast Density",
    yaxis_title="Percentage within Density Category",
    legend_title="BI-RADS Score",
    bargap=0.2,
)

fig.show()
```

```python
import plotly.graph_objects as go

shape_labels = {1: 'Round', 2: 'Oval', 3: 'Lobular', 4: 'Irregular'}
margin_labels = {1: 'Circumscribed', 2: 'Microlobulated', 3: 'Obscured', 4: 'Ill-defined', 5: 'Spiculated'}

grouped_dataheat = mmgData_notnan_filtered.groupby(['Shape', 'Margin'])['Severity'].mean().reset_index(name='mean_severity')

heatmap_data = grouped_dataheat.pivot(index='Shape', columns='Margin', values='mean_severity').fillna(0)

fig = go.Figure(data=go.Heatmap(
    z=heatmap_data.values,
    x=heatmap_data.columns,
    y=heatmap_data.index,
    colorscale='Viridis', 
    colorbar=dict(title='Mean Severity'),
    hoverongaps=False
))

fig.update_layout(
    title='Heatmap of Mean Severity by Shape and Margin',
    xaxis_title='Mass Margin',
    yaxis_title='Shape',
    xaxis=dict(
        tickmode='array',
        tickvals=[1, 2, 3, 4, 5],
        ticktext=[margin_labels[val] for val in [1, 2, 3, 4, 5]]
    ),
    yaxis=dict(
        tickmode='array',
        tickvals=[1, 2, 3, 4],
        ticktext=[shape_labels[val] for val in [1, 2, 3, 4]]
    )
)

# Show the heatmap
fig.show()
```

```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import plotly.express as px

# Assuming mmgData_notnan already exists

# Drop non-numeric or interval columns that can't be used directly
# If 'Age_Range' exists from previous steps, drop it
if 'Age_Range' in mmgData_notnan.columns:
    mmgData_notnan = mmgData_notnan.drop(columns=['Age_Range'])

# Check for categorical columns and encode them (if needed)
categorical_cols = mmgData_notnan.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    mmgData_notnan[col] = LabelEncoder().fit_transform(mmgData_notnan[col])

# Separate features and target variable (assuming 'Severity' is the target)
X = mmgData_notnan.drop(columns=['Age'])
y = mmgData_notnan['Age']

# Apply RFE using Logistic Regression as the estimator
model = LogisticRegression(max_iter=1000)
rfe = RFE(model, n_features_to_select=5)  # Adjust the number of features to select
fit = rfe.fit(X, y)

# Get the ranking of features
ranking = pd.DataFrame({
    'Feature': X.columns,
    'Rank': rfe.ranking_
})

# Plot the ranking of features
fig = px.bar(ranking, x='Feature', y='Rank', title="Feature Ranking by RFE", labels={'Rank': 'RFE Rank (1 = Best)'})
fig.show()

# Output selected features
print("Selected Features by RFE:", X.columns[rfe.support_])
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
import plotly.express as px

X = mmgData_notnan.drop(columns=['Severity'])
y = mmgData_notnan['Severity']

selector = SelectKBest(score_func=chi2, k='all')  # Use 'all' to get scores for all features
X_new = selector.fit_transform(X, y)

# Get feature scores and feature names
scores = selector.scores_
feature_names = X.columns

feature_scores = pd.DataFrame({
    'Feature': feature_names,
    'Score': scores
}).sort_values(by='Score', ascending=False)

fig = px.bar(feature_scores, x='Feature', y='Score', title="Feature Importance Using Chi-Squared Test")
fig.show()

print("Top features based on Univariate Selection:")
print(feature_scores)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import plotly.express as px

# Assuming mmgData_notnan already exists
# Separate features and target variable (assuming 'Severity' is the target)
X = mmgData_notnan.drop(columns=['Severity'])
y = mmgData_notnan['Severity']

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce dimensions
pca = PCA(n_components=2)  # Keep only the first two principal components for visualization
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for the PCA result and add the target variable 'Severity'
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Severity'] = y

# Plot the first two principal components
fig = px.scatter(pca_df, x='PC1', y='PC2', color='Severity',
                 title='PCA of Mammographic Data (First Two Principal Components)',
                 labels={'PC1': 'Principal Component 1', 'PC2': 'Principal Component 2'})
fig.show()

# Print the explained variance ratio
print("Explained variance ratio of each component:", pca.explained_variance_ratio_)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming mmgData_notnan already exists
# Separate features and target variable (assuming 'Severity' is the target)
X = mmgData_notnan.drop(columns=['Severity'])

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Keep the first two components
pca.fit(X_scaled)

# Get the principal component loadings
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=X.columns)

# Show the loadings for each feature
print("Principal Component Loadings for PC1 and PC2:")
print(loadings)

# Find the two features with the highest absolute loadings for PC1 and PC2
most_influential_features = loadings.abs().idxmax()
print("\nMost influential features for PC1 and PC2:")
print(most_influential_features)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import plotly.express as px

# Assuming mmgData_notnan already exists
# Separate features and target variable (assuming 'Severity' is the target)
X = mmgData_notnan.drop(columns=['Severity'])
y = mmgData_notnan['Severity']

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Number of components to match the number of selected features in Task 4.1
# For this example, we assume 5 features were selected
num_components = 5  # Replace with the actual number from Task 4.1

# Apply PCA with the selected number of components
pca = PCA(n_components=num_components)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for the PCA result and add the target variable 'Severity'
pca_columns = [f'PC{i+1}' for i in range(num_components)]
pca_df = pd.DataFrame(data=X_pca, columns=pca_columns)
pca_df['Severity'] = y

# Plot the first two principal components for visualization (optional)
fig = px.scatter(pca_df, x='PC1', y='PC2', color='Severity',
                 title=f'PCA of Mammographic Data ({num_components} Principal Components)',
                 labels={'PC1': 'Principal Component 1', 'PC2': 'Principal Component 2'})
fig.show()

# Print the explained variance ratio for each component
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio of each component:", explained_variance)

# Cumulative explained variance
cumulative_variance = explained_variance.cumsum()
print("Cumulative explained variance:", cumulative_variance)

# Optionally plot the explained variance ratio and cumulative variance
import plotly.graph_objects as go

fig_variance = go.Figure()
fig_variance.add_trace(go.Scatter(x=list(range(1, num_components+1)), y=cumulative_variance, mode='lines+markers', name='Cumulative Variance'))
fig_variance.add_trace(go.Bar(x=list(range(1, num_components+1)), y=explained_variance, name='Explained Variance'))

fig_variance.update_layout(title=f'Explained Variance by {num_components} Principal Components',
                           xaxis_title='Principal Component',
                           yaxis_title='Variance Explained')
fig_variance.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Assuming mmgData_notnan already exists
# Separate features and target variable (assuming 'Severity' is the target)
X = mmgData_notnan.drop(columns=['Severity'])

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA with 2 components
pca = PCA(n_components=2)
pca.fit(X_scaled)

# Get the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)

# Get the principal components
first_pc = pca.components_[0]
second_pc = pca.components_[1]

# Transform the data to the PCA space
transformed_data = pca.transform(X_scaled)

# Plot the first two principal components with original data points
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c='blue', label='Transformed data')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA Transformation with 2 Components")
plt.legend()
plt.show()

# Output: explained variance ratio
print(f'Explained variance of first two components: {explained_variance}')
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming mmgData_notnan already exists
# Separate features and target variable (assuming 'Severity' is the target)
X = mmgData_notnan.drop(columns=['Severity'])

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA with 3 components
pca = PCA(n_components=3)
pca.fit(X_scaled)

# Get the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)

# Get the principal components
first_pc = pca.components_[0]
second_pc = pca.components_[1]
third_pc = pca.components_[2]

# Transform the data to the PCA space
transformed_data = pca.transform(X_scaled)

# Plot the first three principal components in 3D space
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(transformed_data[:, 0], transformed_data[:, 1], transformed_data[:, 2], c='blue', label='Transformed data')
ax.set_xlabel("First Principal Component")
ax.set_ylabel("Second Principal Component")
ax.set_zlabel("Third Principal Component")
plt.title("PCA Transformation with 3 Components")
plt.legend()
plt.show()

# Output: explained variance ratio
print(f'Explained variance of first three components: {explained_variance}')
```

```python
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Assuming mmgData_notnan already exists
# Separate features and target variable (assuming 'Severity' is the target)
X = mmgData_notnan.drop(columns=['Severity'])

# Standardize the features (important for SVD)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply TruncatedSVD
k = 2  # You can change k to the number of components you want
svd = TruncatedSVD(n_components=k)
X_svd = svd.fit_transform(X_scaled)

# Adjust the x-axis to show the actual number of components
plt.plot(range(1, k + 1), np.cumsum(svd.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.title('Truncated SVD Explained Variance')
plt.show()


# Print explained variance ratio for each component
print("Explained variance ratio for each component:", svd.explained_variance_ratio_)
```
