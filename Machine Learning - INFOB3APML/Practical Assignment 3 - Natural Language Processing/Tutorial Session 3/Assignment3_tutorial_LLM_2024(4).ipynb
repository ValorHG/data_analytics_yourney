## Model Training

```python
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))['data']

model = BERTopic(language="english", calculate_probabilities=True, nr_topics="auto", embedding_model="all-MiniLM-L6-v2")
```

```python
topics, probs = model.fit_transform(docs)
```

## Question: data preprocessing?

### Example: stop words

Stop words can sometimes appear in topic representations, but they are generally undesirable as they add little interpretive value. However, removing stop words during preprocessing is not recommended because transformer-based embedding models rely on the full context for accurate embeddings.

A better approach is to use `CountVectorizer` to preprocess documents after generating embeddings and clustering them.

```python
model.get_topic_info()[1:11]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words="english")
model.update_topics(docs, vectorizer_model=vectorizer_model)
```

```python
model.get_topic_info()[1:11]
```

Take-away:
- Removing stop words is more critical in traditional bag-of-words models or static word embeddings (e.g., Word2Vec, GloVe) to reduce dimensionality and noise.
- However, when using transformer-based models (e.g., BERT, GPT, RoBERTa) to obtain word embeddings, removing stop words is generally not necessary. 
- Transformer models generate contextualized embeddings, meaning each word’s representation depends on the surrounding context. Even stop words like “the,” “is,” or “in” are assigned meaningful embeddings based on their role in the sentence.
- Removing stop words can disrupt the context and alter the embeddings of the remaining words. Removing stop words before feeding text to a transformer might leave “gaps” in the sequence, altering the sentence structure and context representation.
